```python
from flair.data import Sentence
from flair.embeddings import WordEmbeddings

embedding = WordEmbeddings('crawl')
sentence = Sentence('one two three one')
embedding.embed(sentence)

for token in sentence:
    print(token.embedding)
```

```python
token1 = sentence[0]
token2 = sentence[3]

print(token1.embedding.tolist() == token2.embedding.tolist())
```

Since classic word embeddings are non-contextual (also no preprocessing required) the first and last token of the sentence, which is 'one', have the same embeddings.
