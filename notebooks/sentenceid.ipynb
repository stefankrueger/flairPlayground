```python
import csv
import pandas as pd
import numpy as np

filepath = '../data/ler/ler.conll'
```

```python
df = pd.read_csv(filepath, sep=' ', header=None, names=['token','tag'], quoting=csv.QUOTE_NONE,skip_blank_lines=False)
```

```python
#66730 NaN tokens im Datensatz, Am Ende ist noch ein NaN, ich bekomme 66729 Sätze raus, Leitner behauptet 66723 Sätze
check_nan_all = df['token'].isnull().sum()
print(check_nan_all)
```

```python
# a bit dirty, but first create col sentence_id with either 
# 0 value  for a non NaN data row
# -1 value for a NaN data row
df['sentence_id'] = df['token'].apply(lambda x: 0 if not pd.isnull(x) else -1)
df['sentence_id'].astype(int) # else sentence_id would be float
df.dtypes
```

```python
check_nan = df['token'].isnull().values.any()
print (check_nan)
```

```python
def build_sentence_ids():
    ID = 0
    last_one_was_minus_one = False
    for index, value in df['sentence_id'].items():
        if(not last_one_was_minus_one) and df.at[index,'sentence_id'] != -1:
            df.at[index,'sentence_id'] = ID
            last_one_Was_minus_one=False
        elif(last_one_was_minus_one):
            df.at[index,'sentence_id'] = ID
            last_one_was_minus_one=False
        else:
            last_one_was_minus_one=True
            ID = ID + 1
```

```python
# 1m41s  xD
build_sentence_ids()
df.head(50)
```
